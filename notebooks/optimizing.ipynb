```python
# Importações

import pandas as pd


from src.config import CLEANED_PATH
```

```python
# Obtendo a base de dados
df = pd.read_parquet(CLEANED_PATH)

df.head()
```

```python
# Definição de cada tipo de coluna

target_column = pd.Index(["median_house_value"]).tolist()

categoric_columns = df.select_dtypes(["object", "category"]).columns.tolist()

numeric_columns = df.select_dtypes("number").columns.drop(target_column).tolist()
```

```python
# Resumo da base
df.info()
```

```python
df["ocean_proximity"].dtype == 'O'
```

```python
# Transformação das colunas object para category

for column in categoric_columns:

    if df[column].dtype == 'O':

        df[column] = pd.Categorical(df[column], ordered=False)
```

```python
df.info()
```

```python
# Fazendo a coerção de colunas numéricas do tipo float64 para int (8, 16 ou 32 bits) e float32

for column in numeric_columns + target_column:

    if df[column].apply(float.is_integer).all():

        df[column] = pd.to_numeric(df[column], downcast="integer")
        continue

    df[column] = pd.to_numeric(df[column], downcast="float")
```

```python
df.info()
```

```python
# Resumo estatístico da base (Para analisar inconsistências)
df.describe()
```

```python
# Exportando a base
df.to_parquet(CLEANED_PATH, index=False)
```
