```python
# Importações

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


from src.config import HOUSING_PATH, CLEANED_PATH
from src.graphics import compositionHistBox
```

```python
# Obtendo a base de dados
df = pd.read_csv(HOUSING_PATH, compression="zip")

df.head()
```

```python
# Definição de cada tipo de coluna

target_column = pd.Index(["median_house_value"]).tolist()

categoric_columns = df.select_dtypes("object").columns.tolist()

numeric_columns = df.select_dtypes("number").columns.drop(target_column).tolist()
```

```python
# Resumo da base
df.info()
```

```python
# Resumo estatístico da base
df.describe()
```

## Colunas numéricas

```python
# Boxplot e Histograma

compositionHistBox(df, numeric_columns + target_column)
```

```python
# Verificando a assimetria
df[numeric_columns + target_column].skew()
```

```python
# Verificando a curtose
df[numeric_columns + target_column].kurtosis()
```

## Colunas Categóricas

```python
# Verificando a contagem de ocorrencias nas colunas categóricas
for column in categoric_columns:

    display(df[column].value_counts())
```

```python
fig, axs = plt.subplots(nrows=len(categoric_columns), ncols=1, figsize=(10, 5))

if not isinstance(axs, np.ndarray):

    axs = np.array([axs])

for ax, column in zip(axs, categoric_columns):

    print(df[column].value_counts().index)

    sns.barplot(
        x=df[column].value_counts().index,
        y=df[column].value_counts().values,
        ax=ax
        )

plt.show()
```

Há poucas ocorrências de `ISLAND` em `ocean_proximity`. Talvez seja interessante remover os registros que sejam dessa categoria.

```python
# Removendo a classe "ISLAND" de "ocean_proximity"
df = df[df["ocean_proximity"] != "ISLAND"]
```

```python
# Exportando a base tratada
df.to_parquet(CLEANED_PATH, index=False)
```
